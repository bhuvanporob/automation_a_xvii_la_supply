```python
import pandas as pd 
```

```python
df1= pd.read_csv('D:/automation_a_xvii_la_supply/dashboard files/5 to 17 feb data raw.csv')
df1.head()
```

```python

# from python file named clean_data
# Cleaning steps
df_wun = df1[~df1['User_name'].isin(['Service'])]
df_wun = df_wun[~df_wun['Lab_name'].isin(['QC'])]
df_clean = df_wun.assign(Truelab_id=df_wun['Truelab_id'].apply(lambda x: x.partition('-')[0]))
df_clean['Chip_serial_no'] = df_clean['Chip_serial_no'].str[:2]
df_clean = df_clean.drop_duplicates(keep='first')
df_clean['Ct1'] = pd.to_numeric(df_clean['Ct1'], errors='coerce').fillna(0)
df_clean['Ct2'] = pd.to_numeric(df_clean['Ct2'], errors='coerce').fillna(0)
df_clean['Ct3'] = pd.to_numeric(df_clean['Ct3'], errors='coerce').fillna(0)
df_clean.head()
```

```python
df_clean.info()
```

```python
#Pivot for MTB Test Status and using lambda function calculate INV per bad
pivot = pd.pivot_table(df_clean, index= ['Lot','Chip_serial_no'], values='Patient_id', columns='Test_status', aggfunc= 'count',margins=True)
pivotdf = pd.DataFrame(pivot.to_records())
# Specify the columns to fill NaN with 0
columns_to_fill = ['Error-1','Error-1A','Error-2','Error-3','Error-4','Error-5','Invalid']

# Fill NaN values in the specified columns with 0
pivotdf[columns_to_fill] = pivotdf[columns_to_fill].fillna(0)
pivotdf["INV_per"] = round(pivotdf.Invalid/pivotdf.All*100,2)
#pivot1df = round(pivot1df.assign(INV_per = lambda x: x.Invalid / x.All * 100),2)

pivotdf.head(20)
```
